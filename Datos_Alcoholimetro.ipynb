```python
import pandas as pd

encuestados = pd.read_csv( "https://raw.githubusercontent.com/Mat3eo56/ProyectoFinalProgramacion/main/Sujetos%20PruebaComa.csv", index_col=0)

encuestados.iloc[:]
```

```python
encuestados.columns= ['n_semanas','salud','c_dinero','prueba1', 'prueba2']
encuestados.iloc[:]
```

```python
promedio = encuestados.n_semanas.mean()
print(f"Promedio:{promedio}")

total_sem = sum(encuestados.n_semanas) 
porcentaje_Andres= ((encuestados.iloc[0].n_semanas)/total_sem)
porcentaje_Guillermo= ((encuestados.iloc[1].n_semanas)/total_sem)
porcentaje_Daniel= ((encuestados.iloc[2].n_semanas)/total_sem)


print(f"Porcentaje Andres: {porcentaje_Andres*100}")
print(f"Porcentaje Guillermo: {porcentaje_Guillermo*100}")
print(f"Porcentaje Daniel: {porcentaje_Daniel*100}")

```

```python
dinero_s_Andres = (encuestados.iloc[0].c_dinero)*(encuestados.iloc[0].n_semanas)
dinero_s_Guillermo= (encuestados.iloc[1].c_dinero)*(encuestados.iloc[1].n_semanas)
dinero_s_Daniel = (encuestados.iloc[2].c_dinero)*(encuestados.iloc[2].n_semanas)

dinero_m_Andres = dinero_s_Andres*4
dinero_a_Andres = dinero_m_Andres*12

dinero_m_Guillermo = dinero_s_Guillermo*4
dinero_a_Guillermo = dinero_m_Guillermo*12

dinero_m_Daniel = dinero_s_Daniel*4
dinero_a_Daniel = dinero_m_Daniel*12
lista_anual= [dinero_a_Andres,dinero_a_Guillermo,dinero_a_Daniel]
encuestados["Dinero_Anual"] = lista_anual

print(encuestados.iloc[:])
print("")

def dinero_anual_gastado():
  for i in lista_anual:
    if i > 6_500_000:
      print(f"Con el dinero gastado, se puedo haber pagado un semestre en la Universidad EAN, No siga gastando dinero en alcohol")
    else:
      print(f"No siga gastando dinero en alcohol")

print(dinero_anual_gastado())

```
